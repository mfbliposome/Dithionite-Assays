```python
# Define the columns to keep
columns_to_keep = ["Unnamed: 0", "31 Trial 1", "31 Trial 2 ", "31 Trial 3", 
                   "31 Triton Trial 1", "31 Triton Trial 2 ", "31 Triton Trial 3"]

# Keep only the selected columns
df_filtered = df_1[columns_to_keep]

# Rename 'Unnamed: 0' to 'time' for clarity
df_filtered = df_filtered.rename(columns={"Unnamed: 0": "time"})
df_filtered = df_filtered.iloc[1:].reset_index(drop=True)


# Display the first few rows
df_filtered.head()
```

### Start modeling

```python
import scipy.stats as stats
```

##### First-Order Kinetics
- $I(t)=I_0e^−kt$
##### Linear equation
- $lnI(t)=lnI_0−kt$

```python
# Select time and fluorescence columns
time = df_filtered["time"].astype(float)
fluorescence_cols = ["31 Triton Trial 1", "31 Triton Trial 2 ", "31 Triton Trial 3"]
fluorescence = df_filtered[fluorescence_cols].astype(float)

# Compute natural log of fluorescence
ln_fluorescence = np.log(fluorescence)

# Fit a linear regression for each trial
rates = {}
plt.figure(figsize=(8, 6))
for col in fluorescence_cols:
    slope, intercept, r_value, p_value, std_err = stats.linregress(time, ln_fluorescence[col])
    rates[col] = -slope  # Rate constant k
    plt.plot(time, ln_fluorescence[col], 'o', label=f'{col} Data')
    plt.plot(time, intercept + slope*time, '--', label=f'{col} Fit (k={-slope:.4f})')

# Display results
plt.xlabel("Time")
plt.ylabel("ln(Fluorescence Intensity)")
plt.legend()
plt.title("Reaction Rate Estimation from Fluorescence Decay")
plt.show()

# Print reaction rates
print("Estimated reaction rates (k):")
for col, rate in rates.items():
    print(f"{col}: {rate:.4f}")
```

```python
r_value, p_value, std_err
```

```python
# Define an early time window for fitting 
time_cutoff = 100
time_early = time[:time_cutoff]
ln_fluorescence_early = ln_fluorescence.iloc[:time_cutoff]

# Fit only early decay points
rates = {}
plt.figure(figsize=(8, 6))

for col in fluorescence_cols:
    slope, intercept, r_value, p_value, std_err = stats.linregress(time_early, ln_fluorescence_early[col])
    rates[col] = -slope  # Reaction rate constant k
    
    # Plot actual data
    plt.plot(time, ln_fluorescence[col], 'o', label=f'{col} Data', alpha = 0.3)
    
    # Plot fitted line using only the initial decay for fit
    plt.plot(time_early, intercept + slope*time_early, '--', label=f'{col} Fit (k={slope:.4f})')

# Display results
plt.xlabel("Time")
plt.ylabel("ln(Fluorescence Intensity)")
plt.legend()
plt.title("Reaction Rate Estimation from Fluorescence Decay")
plt.show()
```

```python
print("Estimated reaction rates (k):")
for col, rate in rates.items():
    print(f"{col}: {-rate:.4f}")
```
